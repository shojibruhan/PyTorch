# Question 01: [ Marks 10 ]
Explain gradient descent in detail. What is the main idea behind it? How does it help in minimizing the loss function? Explain the role of learning rate and describe what happens if the learning rate is too high or too low.

**Write** Answer 01:

**Gradient Descent**
Gradient Descent is an optimization algorithm which use to minimize the loss function i.e error by updating bias and weights.

Basically, in linear regression we try to find best fit line by updating m(slope) and b (bias) of

$$ y= mx + b $$

It's easy to achive with the help of gradient descent. We could achive it through perceptron learning rule but there are some drawback. It can fails many times and also some time it can't converge.

Main idea of gradient descent is that find the optimum value of weight and bias that can minimize the loss function for better accuracy. Usually model calculate gradient of the loss, update the loss with the help of learning rate(α) and repeat the process until convergence.

Loss function:

$$ L(m, b)= \sum_{i=n}^{n}(y - mx_i - b)^2 $$

**Role of Learning Rate (η)**

Learning rate is the speed of the model's learning state. That means how fast or slow model should learn. It is the step size taken towards the loss functions.

If it gets too high, model get confused i.e it oscillate or diverge.

If it gets too low, it takes forever to converg or in the worst case it gets stuck.

# Question 04: [ Marks 15 ]

Explain the sigmoid activation function. Where is it used? What are its advantages and disadvantages? What problem does it cause during training?

## Write Answer 04:

Sigmoid function is a very basic activation function which convert any value between 0 and 1. It is used in machine learning, deep learning i.e neural networks for modeling binary classification problems and introducing non-lenearity into models. This is the resone why it use in logistic regression.


$$ \sigma(x)=  \frac{1}{1+e^{-x}} $$
Here,
$$ x= input features $$
$$ σ(x)= output \ value $$
$$e= Euler's \ numbe(≈ 2.718)$$

> For negetive values : (x → - ∞) output is 0.
>
> For positive values: (x → + ∞) output is 1.

<img src="https://cdn.britannica.com/64/264764-050-A2C174FD/graph-of-a-sigmoid-function.jpg" width=400/>


**Advantage**
* Ideal for probablity prediction, binary classification.
* Clearly separate classes into 0 or 1.
* Complete analytical derivative.
* Produce somooth, globally differentiable outputs


**Disadvantage**
* Output is non zero centerd. As a result, there is a chance of bias shift during optimaization.
* It can lead to vanishing gradient problem where gradient becomes so small that it stops learning.
* In saturate regions, gradient collapes.
* Getting slow when the magnitude of inputs get larger.


**Problem creates during training**


Primarily it creates `vanishing gradient problem` during training. In simple terms gradient getting small- shrinks to 0. So model try to to update the weight with extremly small number resulting in very slow learning. Eventually model stop to learn.


# Question 05: [ Marks 10 ]
Explain the softmax activation function in simple words. Why is it used in multi-class classification problems? How is it different from sigmoid?

## Write Answer 05:

**Softmax Activation Function**
Softmax activation is a layer which designated to solve multiclass classification. When we have to choose one class out of many we use softmax activation function. It trasforms the logit to a normalized problity distribution.

$$ σ(z_i)= \frac{e^{z_i}}{\sum\limits_{j=1}^{k} e^{z_j}} $$

Let's say value of z:
$$ z= [2, 1, 0.1] $$
So,

$$ e^2= 7.39, \ \  e^1= 2.72, \ \  e^{0.1}= 1.105 $$
Total sum:
$$ 7.39+ 2.72+ 1.105= 11.215 $$

deviding the each exponential value with sum:

$$ [\frac{7.39}{11.215}, \frac{2.72}{11.215},  \frac{1.105}{11.215}] $$

$$ [0.659, 0.242, 0.099] $$

This is actually probality distribution.

**Why Softmax is used in multi-class classification**

* It transforms raw scores into probablities where the range is between [0,1]

* Ensure total sum of the output is exactly 1. It is necessary for choosing winning class. That is if probablity of one class goes up others go down.

* It highlights the largest value and minimize the impact of smaller value. It helps to create distinct prediction.


**Softmax activation Vs Sigmoid**

Sigmoid activation is used in binary classification. It map any real value input into exactly 0 or 1. Sigmoid treat every output independently. But it can't use in mult-class classification. This is why we use softmax activation.It is used for multi-class classification.

🧠 **Artificial Neural Network (ANN)**

Artificial Neural Network in short ANN is a branch of Machine learning. It inspired from human neuron to learn pattern and make decission from given input.

🕸️ **Main components**

* <u>**Neuron or Node** </u>  
It's the smallest unit of ANN which take input, process the data and yield an output.
$$ y= f(\sum_{i=1}^{n}w_ix_i + b ) $$
Where,  
y = output,  
f= activation function,  
wᵢ= weight,

b= bias

  2. Hidden Layers: Process data from input
  3. Output Layers: Yield output

* <u>Weights & Bias</u>  

    ⦿ Weights:  Control the influence of the inputs.  
    ⦿ Bias: Allow flexibility by shifting the activation function.
* <u>Activation Function</u>

  ⦿ Adds non-linearity. Helps to learn complex pattern. example are RelU, sigmoid, Tanh etc.

ANN update the weights to make more accurate prediction which is called `Backward Pass`. For activation it uses `Gradient Descent Algorithm`.
